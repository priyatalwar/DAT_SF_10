```python
json_data = open('ga_hw_logins.json')
```

```python
data = json.load(json_data)
data
```

```python
# convert the string into datetime objects and append them to a list
ts_data = []
```

```python
for item in data:
    ts_data.append(datetime.datetime.strptime(item, '%Y-%m-%d %H:%M:%S'))
```

```python
# import datetime lib
import datetime
ts_data
```

```python
# import sqlite 
import sqlite3
```

```python
# create a sqlite3 database andupload the data into the database
con = sqlite3.connect('hw1.db')
```

```python
cur = con.cursor()
try:
    cur.execute("drop table hw1")
except:
    print 'dropping failed because there may not have been a table.'
cur.execute("create table hw1(ts timestamp)")
for data in ts_data:
    cur.execute("insert into hw1(ts) values(?)", (data,))
ts_data
```

```python
# Query the data to find the date and hour with the most logins
cur.execute("select strftime('%H', ts) FROM hw1")
print cur.fetchone()
cur.execute("select count(*)c, strftime('%H', ts) hour FROM hw1 group by hour order by c desc limit 5;")
print cur.fetchone()
cur.execute("select count(*) c, strftime('%Y-%m-%d', ts) date FROM hw1 group by date order by c desc limit 1;")
print cur.fetchone()
```
