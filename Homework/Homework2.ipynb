```python
# KFold cross validation
from sklearn.cross_validation import KFold
```

```python
print X.shape, y.shape
```

```python
def cross_validate(X, y, classifier, k_fold) :

    # derive a set of (training and testing indices
    k_fold_indices = KFold( len(X), n_folds=k_fold,
                           indices=True, shuffle=True)

    k_score_total = 0
    # for each training and testing slices run the classifier, and score the results
    for train_slice, test_slice in k_fold_indices :

        model = classifier(X[ train_slice  ],
                         y[ train_slice  ])

        k_score = model.score(X[ test_slice ],
                              y[ test_slice ])

        k_score_total += k_score

    # return the average accuracy
    return k_score_total/k_fold
```

```python
# using 5 folds for cross validation 
cross_validate(X, y, KNeighborsClassifier(5).fit, 5)
```

```python
# To find optimal value of k, we need to loop over possible values of K and then find K for which average accuracy is maximized
knn_values = np.array(range(1,100))
knn_results = []
for k in range(1,100):
    knn_results.append(cross_validate(X, y, KNeighborsClassifier(k).fit, 5))

# for k in range(1,100) gives an "out of bound error- may be a python error?"
```
