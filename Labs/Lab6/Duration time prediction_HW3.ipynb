```python
data = pd.read_csv('student_logins.csv')
```

```python
data['session_start_time']=pd.to_datetime(data['session_start_time'])
data['session_end_time']=pd.to_datetime(data['session_end_time'])
data['session_duration']=pd.to_datetime(data['student_account_created'])
data['student_account_created']=pd.to_datetime(data['student_account_created'])
```

```python
data.info()
```

```python
# Transforming into secs
data['session_duration'] = (data.session_end_time - data.session_start_time) / np.timedelta64(1, 's')
```

```python
# Adding column and trasnforming into Days - alternative method
data['age_account_creation']=pd.to_datetime(data['session_start_time']) - pd.to_datetime(data['student_account_created'])
data['age_account_creation'] = data['age_account_creation'].astype('timedelta64[D]')
```

```python
# creating a dictionary and adding a column with a numerical index 
class_id_dict = {class_id : i for i, class_id in list(enumerate(np.unique(data['class_id']))) }
class_id_dict
```

```python
data['class_id_dict'] = data['class_id'].map(lambda x: class_id_dict[x])
data.head()
```

```python
# use dummy valriables in pandas
# use cross validation or fitted R 2 or mean sq. error for the model
# forward selection iterates through each feature
# create features based on students, depending on how fast they are
```

```python
# Running Regression
x = data[['problems_completed', 'student_previous_logins_total', 'student_previous_class_logins', 'student_id', 'age_account_creation', 
            'class_id_dict']].values
y = data[['session_duration']].values
X = sm.add_constant(x, prepend=True)
results = sm.OLS(y, X).fit()
results.summary()
```

```python
# Adjusted R sq. is .415 and let's see if we can improve it by seeing which variables are significant
x = data[['problems_completed', 'student_previous_class_logins', 'student_id', 'age_account_creation', 
            'class_id_dict']].values
y = data[['session_duration']].values
X = sm.add_constant(x, prepend=True)
results = sm.OLS(y, X).fit()
results.summary()
```
